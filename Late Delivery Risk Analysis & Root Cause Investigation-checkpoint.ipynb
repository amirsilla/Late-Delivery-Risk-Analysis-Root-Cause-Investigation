# Supply Chain Performance Analysis

## Project Overview

This project analyzes the DataCo Supply Chain dataset using Python and Pandas to evaluate sales performance, profitability, logistics efficiency, and product performance.

## Key Metrics

* Total Sales: 36.78 Million
* Total Profit: 3.97 Million
* Total Orders: 65,752
* Average Shipping Time: 3.50 Days
* Late Delivery Risk: 54.83%

## Key Findings

### Category Performance

* Fishing generated the highest sales (6.93M) and highest profit (756.22K).
* Cleats generated the second-highest sales (4.43M) and profit (494.64K).
* Camping & Hiking was the third best-performing category with sales exceeding 4.11M.

### Regional Performance

* Western Europe generated the highest sales (5.89M).
* Central America ranked second with sales of 5.67M.
* South America generated approximately 2.96M in sales.
* Western Europe also produced the highest profit (625.45K).

### Shipping Performance

* Same Day shipping was the fastest shipping mode with an average delivery time of 0.48 days.
* First Class averaged 2.00 days.
* Standard Class averaged 4.00 days.

### Delivery Risk Analysis

* First Class showed the highest late delivery risk (95.32%).
* Second Class recorded a late delivery risk of 76.63%.
* Standard Class had the lowest late delivery risk (38.07%).

### Product Performance

* Field & Stream Sportsman 16 Gun Fire Safe generated the highest sales (6.93M).
* Perfect Fitness Perfect Rip Deck ranked second in sales performance.
* Diamondback Women's Serene Classic Comfort Bike ranked third.

## Business Recommendations

1. Increase inventory availability for high-performing categories such as Fishing, Cleats, and Camping & Hiking.
2. Investigate the extremely high late delivery risk in First Class shipping.
3. Expand marketing efforts in Western Europe and Central America due to strong sales performance.
4. Review logistics operations to reduce delivery delays and improve customer satisfaction.
5. Prioritize high-revenue products in inventory planning and forecasting.
